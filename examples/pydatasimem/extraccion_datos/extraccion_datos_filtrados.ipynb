## Objetivos de este Notebook

* Explicar cómo se puede consultar el catálogo de conjuntos de datos que se publican en SiMEM.
* Explicar cómo consultar la información de un conjunto de datos determinado.
* Demostrar cómo hacer filtros especializados en un conjunto de datos determinado.

### 1. Instalación

```python
from pydataxm.pydatasimem import ReadSIMEM, VariableSIMEM, CatalogSIMEM
```

### 2. Catálogos

##### CatalogSIMEM

```python
catalogo_dataset = CatalogSIMEM(catalog_type='Datasets').get_data()
catalogo_dataset.head()
```

```python
nombre_dataset = 'Precio de Bolsa'

idDataset = catalogo_dataset[
    catalogo_dataset['nombreConjuntoDatos']
        .str.lower()
        .str.contains(nombre_dataset.lower(), na=False)
]

idDataset[['idDataset', 'nombreConjuntoDatos']]
```

```python
catalogo_variables = CatalogSIMEM(catalog_type='Variables').get_data()
catalogo_variables.head()
```

```python
nombre_variable = 'Precio de Bolsa'

idDataset = catalogo_variables[
    catalogo_variables['nombreVariable']
        .str.lower()
        .str.contains(nombre_variable.lower(), na=False)
]

idDataset[['idDataset', 'nombreDataset', 'nombreVariable']]
```

##### VariableSIMEM

```python
listado_variables_simem = VariableSIMEM.get_collection()
listado_variables_simem.head()
```

```python
nombre_variable = 'Precio de Bolsa'

cod_variable = listado_variables_simem[
    listado_variables_simem['Nombre']
        .str.lower()
        .str.contains(nombre_variable.lower(), na=False)
]

cod_variable[['CodigoVariable', 'Nombre']]
```

### 3. Consulta de datos

##### ReadSIMEM

```python
id_dataset = 'EC6945'
fecha_inicio = '2026-01-01'
fecha_fin = '2026-02-28'

simem = ReadSIMEM(dataset_id=id_dataset, start_date=fecha_inicio, end_date=fecha_fin)
```

```python
data = simem.main()
data
```

```python
filtro = [["CodigoVariable","=","PB_Nal"],["Version","=","TXF"],["Valor","between",["95","150"]]]
simem.set_filter(filters=filtro)
```

```python
simem.get_filters()
```

```python
data_filtro = simem.main(filter=True)
data_filtro
```

##### VariableSIMEM

```python
start_date = '2026-01-01'
end_date = '2026-02-28'
codigoVar = 'PB_Nal'
filtro = ["Valor",">=","120"]

data_var = VariableSIMEM(start_date=start_date, end_date=end_date, 
                         cod_variable=codigoVar, filters=filtro).get_data()
data_var
```
